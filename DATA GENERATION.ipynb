```python
# but in real life data is normally distributed as compared to uniform distribution
x=np.random.randn(3000)  # returns the sample of data from normal distribution
```

```python
print(x)
```

```python
plt.hist(x)
plt.show()
```

```python
# if we want to add some mean value to this normal distribution having mean as zero then we can add that value to the data and the datat is shifted accordingly
```

```python
x=np.random.randn(30000)+50 # now the men i shifted to 50 and the distribution is normally distributed with mean 50.
plt.hist(x)
plt.show()
```
