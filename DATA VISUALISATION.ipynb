# DATA VISUALISATION

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
print(x)
```

```python
plt.plot(x) # returns object
```

```python
plt.plot(x)
plt.show()
x2=x**3
```

```python
x=x**2
plt.plot(x)
plt.show()
```

```python
plt.plot(x2,x)
plt.show()
```

```python
i=np.arange(10)
isq=i*i
plt.plot(i,isq)
plt.show()
```

```python
x=np.arange(10)
plt.scatter(x,x*2,color='red',marker='^',label='stocks label')
plt.legend()
plt.xlabel("days")
plt.ylabel("price")
plt.title("Stock price prediction")
plt.show()
```

```python
x=np.arange(5)
plt.plot(x,x**2,color='red',label='2017')
plt.plot(x,x**3,color='green',label='2018')
plt.xlabel("No of days")
plt.ylabel("prices")
plt.title("stocks prices")
plt.legend()
plt.show()
```

#### BAR CHART

```python
numbers=np.random.randint(0,10,5)
indices=np.arange(5)

print(numbers)
print(indices)
```

```python
plt.bar(indices,numbers)
plt.show()
```

```python
plt.bar(indices,numbers,color='orange')
plt.show()
```

```python
plt.bar(indices,numbers,0.5,color='green')
plt.show()
```

```python
plt.bar(indices,numbers,0.5,color='green',label='gold price')
plt.legend()
plt.show()
```

```python
number1=np.random.randint(0,10,5)
indices1=np.arange(5)
```

```python
plt.bar(indices,numbers,0.5,color='red',label='tomato price')
plt.bar(indices1,number1,0.5,color='green',label='grapes price')
plt.legend()
plt.show()
```

```python
plt.bar(indices,numbers,0.5,color='red',label='tomato price')
plt.bar(indices1+0.5,number1,0.5,color='green',label='grapes price')
plt.legend()
plt.show()
```

```python
plt.bar(indices,numbers,0.5,color='red',label='tomato price')
plt.bar(indices1,number1,0.4,color='green',label='grapes price')
plt.legend()
plt.show()
```

```python
plt.bar(indices,numbers,0.25,color='red',label='tomato price')
plt.bar(indices1+0.2,number1,0.25,color='green',label='grapes price')
plt.legend()
plt.show()
```

#### PIE CHART

```python
labels=['coding','python','algorithms','algebra','calculus']
values=[43,56,444,768,657]
plt.pie(values,labels=labels,radius=1)
plt.show()
```

```python
labels=['coding','python','algorithms','algebra','calculus']
values=[43,56,444,768,657]
plt.pie(values,labels=labels,radius=1,explode=(0,0,1,0,0))
plt.show()
```

```python
labels=['coding','python','algorithms','algebra','calculus']
values=[43,56,444,768,657]
plt.pie(values,labels=labels,radius=1,explode=(0,0,1,0,0),shadow=True)
plt.show()
```
