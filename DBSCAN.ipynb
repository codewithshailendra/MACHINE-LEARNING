```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
x,y = make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
# make object of DBSCAN
dbscan = DBSCAN(eps=0.2,min_samples=5)
```

```python
dbscan.fit(x)
```

```python
ypredict=dbscan.fit_predict(x)

```

```python
print(ypredict)
```

```python
plt.scatter(x[:,0],x[:,1],c=ypredict)
plt.show()
```
