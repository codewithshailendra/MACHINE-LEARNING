```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn')
```

#### Data Preparation

```python
mean_1 = np.array([1,0.5])
cov_1 = np.array([[1,0.1],[0.1,1.2]])

mean_2 = np.array([4,5])
cov_2 = np.array([[1.21,0.1],[0.1,1.3]])


# Normal Distribution

dist_01 = np.random.multivariate_normal(mean_1,cov_1,500)
dist_02 = np.random.multivariate_normal(mean_2,cov_2,500)

print(dist_01.shape)
print(dist_02.shape)
```

#### Data Visualisation

```python
plt.scatter(dist_01[:,0],dist_01[:,1],label ='class 0')
plt.scatter(dist_02[:,0],dist_02[:,1],label='class 1',color='r',marker='^')
plt.xlim(-5,10)
plt.ylim(-5,10)
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```
