#### Gradient Descent Algorithm

#### 1. Start with a random theta 
#### 2. Repeat until convergence
 #### 2.1 Update theta according to our updation rule.

```python
def hypothesis_function(x,theta):  # theta =[theta_not, theta_one], it is an array or two dimentional vector in terms of real algebra.
    return theta[0]+theta[1]*x
```

```python
def total_error(X,Y,theta):        # function to find the total_error
    no_of_samples = X.shape[0]
    error=0
    for i in range(no_of_samples):
        error+=(hypothesis_function(X[i][0],theta)-Y[i][0])**2
        #print(Y[i])
    return error

def gradient_function(X,Y,theta):  # function to find the summation of all the gradients.
    m = X.shape[0]
    gradient = np.zeros((2,))
    for i in range(m):
        gradient[0] += hypothesis_function(X[i],theta)-Y[i][0]
        gradient[1] += (hypothesis_function(X[i],theta)-Y[i][0])*X[i][0]
    return gradient
        
        
```

```python
# Main Algorithms implementation
def gradient_descent(X,Y,learning_rate=0.001):
    theta = np.zeros((2,))
    iteration = 0
    maximum_iteration = 100
    error_list = []
    
    while (iteration<=maximum_iteration):
        gradient = gradient_function(X,Y,theta)
        error = total_error(X,Y,theta)
        #print(error)
        error_list.append(error)
        theta[0] = theta[0] - learning_rate*gradient[0]
        theta[1] = theta[1] - learning_rate*gradient[1]
        iteration+=1
        
    return theta,error_list
     
    
```

```python
optimal_theta,error_array = gradient_descent(X,Y)
```
