```python
from mpl_toolkits.mplot3d import Axes3D # import the required library
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = [1,2,3]
b = [4,5,6]
a,b = np.meshgrid(a,b)

print("a is printing")
print(a)
print("b is printing")
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
a = np.arange(-5,5,0.02)
b = a
a,b = np.meshgrid(a,b)
print("Printing A")
print(a)
print("Printing B")
print(b)
```

```python
m = a.shape[0]
n = a.shape[1]
print(m,n)
```

```python
for i in range(m):
    for j in range(n):
        print("(%d,%d)"%(b[i,j],a[i,j]),end=" ")
    print()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='coolwarm')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```
